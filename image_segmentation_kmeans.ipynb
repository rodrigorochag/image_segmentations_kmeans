```python
import pandas as pd
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import cluster
```

```python
!wget https://www.cnnbrasil.com.br/wp-content/uploads/sites/12/2021/09/pexels-pixabay-46148-e1632575759700.jpg?w=1220&h=674&crop=1
```

```python
image = plt.imread('/content/pexels-pixabay-46148-e1632575759700.jpg?w=1220')
```

```python
def read_image(image):
  plt.axis('off')
  plt.imshow(image)
```

```python
read_image(image)
```

```python
image.shape
```

```python
print(image)
```

```python
vetor_image = np.reshape(image, (814*1220,3))
vetor_image
```

```python
colors_df = pd.DataFrame(vetor_image, columns=['Red', 'Green', 'Blue'])
colors_df
```

```python
image_kmeans = cluster.KMeans(3)
cluster_image = image_kmeans.fit_predict(colors_df)
```

```python
cluster_image = np.reshape(cluster_image, (814,1220))
cluster_image
```

```python
plt.imshow(cluster_image)
plt.axis('off')
plt.show()
```
